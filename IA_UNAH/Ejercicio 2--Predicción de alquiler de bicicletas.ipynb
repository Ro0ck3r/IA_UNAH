```python
import pandas as pd
import numpy as py
#Nos permite importar el tipo de modelo predictivo (regrecion y clasificacion)
from sklearn.model_selection import train_test_split
#para poder hacer uso de la regresion lineal
from sklearn.linear_model import LinearRegression
#para poder verificar cuan presiso fue el resultado obtenido
from sklearn.metrics import mean_squared_error, r2_score
#Nos permite leer un archivo CSV especificando la ruta en cuestion
df_datos = pd.read_csv('C:/users/tavog/day.csv')
df_datos
```

```python
#Verificando la existencia de datos nulos
df_datos.isnull().sum()
```

```python
#dividiendo los datos
X = df_datos[['temp', 'hum', 'windspeed', 'season']]
y = df_datos['cnt']
```

```python
#repartiendo los datos entre las variables
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2)
```

```python
#Crecion de modelo
modelo = LinearRegression()
modelo.fit(X_train, y_train)
```

```python
#Realizar predicciones
y_predic = modelo.predict(X_test)

print("Predicciones por medio del modelo:\n", y_predic) # Valores predecidos en funcion del entrenamiento
print("Valores reales: \n", y_test.values) #Valores realies en Y
```

```python
#Evaluando rendimeinto del modelo
ecm = mean_squared_error(y_test, y_predic)

r = r2_score(y_test, y_predic)
#ECM entre mas alto mejor
print("ECM: ",ecm)
#entre mas cercano a 1 seria mas presiso
print("r: ",r)
```

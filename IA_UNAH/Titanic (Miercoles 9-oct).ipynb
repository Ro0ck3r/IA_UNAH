```python
import pandas as pd
import numpy as py

df_titanic = pd.read_csv('C:/Users/tavog/titanic.csv')
df_titanic
```

```python
#Limpieza de dato
#el metodo .isnull, nos retornaria la cantidad de valores nullos en columna "cabin" en caso que de desear mirar en todo el DT, solo se da lamenra dt.isnull
#El la funcion, sum(), suma cada uno de los valores retornados por el metodo isnull
#print(df_titany.cabin.isnull().sum())
print(df_titanic.isnull().sum())
```

```python
#para limpiar/limpiar datos numericos se deberan ser reemplazados por la media
#Para limpiar/limpiar datos de tipo caracter se deberan reeemplizar con la moda
#La funcion fillana, completa todos aquellos valores nullos para este ejemplo, se reemplaza los valores nullos de la "Age" con la media
#df_titanic.Age.fillna(df_titanic['Age'].mean())
#Metodo 2, haciciendouso del loc
df_titanic.loc[df_titanic['Age'].isnull(), 'Age'] = df_titanic['Age'].mean()
```

```python
print(df_titanic.isnull().sum())
```

```python
#Reemplando valores no validos en el campo Embarked
df_titanic.loc[df_titanic['Embarked'].isnull(), 'Embarked'] = df_titanic['Embarked'].mode()[0]
```

```python
#Reemplando valores no validos en el campo Cabin
df_titanic.loc[df_titanic['Cabin'].isnull(), 'Cabin'] = df_titanic['Cabin'].mode()[0]
print(df_titanic.isnull().sum())
```

```python
#Calcula promedio sobrevivientes y no sobrevivientes
noSurvi = df_titanic.Survived[df_titanic.Survived == 0].count()
survi = df_titanic.Survived[df_titanic.Survived == 1].count()
print("el porcentaje  de supervivientes fue de :", round((survi/df_titanic.Survived.count())*100,2) ,"%" )
print("el porcentaje  de No supervivientes fue de :", round((noSurvi/df_titanic.Survived.count())*100,2) ,"%" )
#Segundo metodo para obtener % de sobrevivientes
print("Este seria la segunda forma de calcular el procentaje de sobrevivientes: ", round(df_titanic.Survived.mean()*100 ,2 ),"%")
print("Este seria la segunda forma de calcular el procentaje de no  sobrevivientes: ", round((1-df_titanic.Survived.mean())*100 ,2 ),"%")
```

```python
#cuÃ¡ndos sobrevivieron de cada sexo
df_survixsex = df_titanic[['Survived','Sex','Age']]
surviM = df_survixsex.Sex[df_survixsex.Sex == "male"].count()
surviF = df_survixsex.Sex[df_survixsex.Sex == "female"].count()
df_surviM = df_survixsex[(df_survixsex['Sex'] == 'male') & (df_survixsex['Survived'] == 1)]
df_surviM.Survived.count()
df_surviF = df_survixsex[(df_survixsex['Sex'] == 'female') & (df_survixsex['Survived']==1)]
df_surviF.Survived.count()
print('Cantidad de sobrevivieron del sexo masculino: ',df_surviM.Survived.count()," representando el ",round((1-(surviM-survi)/survi)*100), "% de los supervivientes")
print('Cantidad de sobrevivieron del sexo femeninos: ',df_surviF.Survived.count()," representando el ",round(((surviM-survi)/survi)*100), "% de los supervivientes")
```

```python
df_survixclass = df_titanic[['Survived','Pclass']]
df_cantidadxClass = df_survixclass.groupby('Pclass').agg('sum')
print("Los sobrevivientes por clase estÃ¡n distribuidos de la siguiente manera:\n" + str(df_cantidadxClass))
```

```python
import pandas as pd
import numpy as py
#Nos permite importar el tipo de modelo predictivo (regrecion y clasificacion)
from sklearn.model_selection import train_test_split
#para poder hacer uso de la regresion lineal
from sklearn.linear_model import LinearRegression
#para poder verificar cuan presiso fue el resultado obtenido
from sklearn.metrics import mean_squared_error, r2_score
#Nos permite leer un archivo CSV especificando la ruta en cuestion
df_datos = pd.read_csv('C:/users/tavog/exams.csv')
df_datos
```

```python
#Verificamos cuantos datos no valido tiene
df_datos.isnull().sum()
```

```python
#Agregamos una columna con la suma de las notas llamada nota final
df_datos['Nota_final'] = ((df_datos['math score'] + df_datos['reading score'] + df_datos['writing score'])/3)
#Aqui creamos una columna y en funcion del valor encontrado en 'gender' agregamos a esa columna un 0 cuando se trate de un hombre y 1 si es mujer
df_datos['genero'] = df_datos['gender'].apply(lambda x: 0 if x == 'male' else 1)
#Aqui creamos una columna y en funcion del valor encontrado en 'race' agregamos a esa columna un 1 cuando sea del grupo A, 2 si es del B , y sino 3 
df_datos['raza'] = df_datos['race/ethnicity'].apply(lambda x: 1 if x == 'group A' else (2 if x=='group B' else (3 if x == 'group C' else (4 if x=='group D' else 5))))
df_datos
```

```python
#partiendo los datos
X = df_datos[['genero','raza','math score', 'reading score', 'writing score']] #Dimensiones
y = df_datos['Nota_final'] #lo que buscamos
```

```python
#Asignamos los datos
#Aqui asignamos el 80% de los datos para pruebas y el 20% para verificar la validez de la predicciones
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
#construimos el modelo
modelo = LinearRegression()
modelo.fit(X_train, y_train)
```

```python
#Realizar predicciones
y_predic = modelo.predict(X_test)

print("prediccones: \n", y_predic) # Valores predecidos en funcion del entrenamiento
print("Valores reales:\n", y_test.values) #valores reales
```

```python
ecm = mean_squared_error(y_test, y_predic)

r = r2_score(y_test, y_predic)
#ECM entre mas alto mejor
print("ECM: ",ecm)
#entre mas cercano a 1 seria mas presiso
print("r: ",r)
```

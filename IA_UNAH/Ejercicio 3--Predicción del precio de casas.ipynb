```python
import pandas as pd
import numpy as py
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
#Ingresandon al archivo csv
df_datos = pd.read_csv('C:/users/tavog/train.csv')
df_datos
```

```python
#Validando los datos
df_datos.TotRmsAbvGrd.isnull().sum() #Esta limpio
df_datos.LotFrontage.isnull().sum() #lo observo que si tiene valores nulos
df_datos.loc[df_datos['LotFrontage'].isnull(), 'LotFrontage'] = df_datos['LotFrontage'].mean() #Limpueza de valores cuando son nulos
df_datos.LotFrontage.isnull().sum() #lo observo que ya no tiene valores nulos
```

```python
#Dividiendo los datos entre las dimensiones/caracteristicas y lo que buscamos predecir (SalePrice)
X = df_datos[['LotArea','TotRmsAbvGrd','LotFrontage']]
y = df_datos['SalePrice']
#Asignando los valores de X y Y , asi como el tamanio de datos con lo que trabajaremos
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2)
```

```python
#Creamos el modelo y lo ponemos a entrenar
modelo = LinearRegression()
modelo.fit(X_train, y_train)
```

```python
#Realizamos predicciones
y_predic = modelo.predict(X_test)

print('Valores predecidos por el modelo: \n', y_predic) # Valores predecidos en funcion del entrenamiento
print("Valores realies: \n", y_test.values) #valores reales
```

```python
#evaluando rendimiento del modelo
ecm = mean_squared_error(y_test, y_predic)

r = r2_score(y_test, y_predic)
#ECM entre mas alto mejor
print("ECM: ",ecm)
#entre mas cercano a 1 seria mas presiso
print("r: ",r)
```

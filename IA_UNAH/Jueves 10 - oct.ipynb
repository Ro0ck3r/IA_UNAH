```python
import pandas as pd
#Nos permite importar el tipo de modelo predictivo (regrecion y clasificacion)
from sklearn.model_selection import train_test_split
#para poder hacer uso de la regresion lineal
from sklearn.linear_model import LinearRegression
#para poder verificar cuan presiso fue el resultado obtenido
from sklearn.metrics import mean_squared_error, r2_score
tamanios = [750, 800, 850, 900, 950, 1000, 1050, 1100, 1150, 1200]
precios = [150000, 160000, 165000, 175000, 185000, 190000, 200000, 210000, 215000, 225000]
datos = {
    'tamanios' : tamanios,
    'precios' : precios 
}
```

```python
df_precios = pd.DataFrame(datos)
df_precios
```

```python
#Al momento de hacer uso de cualquier modelo predictivo (regrecion y clasificacion), deberemos dividir los datos en X y Y
# X Seran todas las feactures sobre las cuales queramos predecir un valor
# Y seran los valores a predecir
X = df_precios[['tamanios']]
y = df_precios['precios']
```

```python
#Aqui estamos partiendo los datos que cuanto % usaremos para entrenamiento y cuentos para pruebas
X_train, X_test, y_train, y_test = train_test_split(X , y , test_size = 0.2 )
```

```python
#Construyendo el modelo
modelo = LinearRegression()
modelo.fit(X_train, y_train)
```

```python
#Realizando predicciones
y_pred = modelo.predict(X_test)

print("Predicciones: ",y_pred) #Valore predecido en funcion del entrenamineto
print("Valores reales: ", y_test.values) # Valores reales
```

```python
ecm = mean_squared_error(y_test, y_pred)

r = r2_score(y_test, y_pred)
#ECM entre mas alto mejor
print("ECM: ",ecm)
#entre mas cercano a 1 seria mas presiso
print("r: ",r)
```
